```python
from astropy import units as u
from astropy.coordinates import SkyCoord, EarthLocation

PIWNICE = EarthLocation(53*u.deg, 18*u.deg)
OBJECTS = None

path = r"C:\Users\Bazik\Desktop\lista_obiektów.txt"

# zakładam, że dostaję plik tekstowy, gdzie:
# nazwa_obiektu ra dec w formie dziesiętnej stopni na ten moment

class Object:
    def __init__(self):
        self.name = None
        self.ra = 0
        self.dec = 0
        self.skycoords = 0

    def FillSkycoords(self):
        self.skycoords = SkyCoord(ra = self.ra*u.deg, dec = self.dec*u.deg)

OBJECTS = []
with open(path, 'r') as file:
    for line in file:
        obj = Object()
        elements = line.split()
        obj.name = str(elements[0])
        obj.ra = float(elements[1])
        obj.dec = float(elements[2])
        obj.FillSkycoords()
        OBJECTS.append(obj)
```
